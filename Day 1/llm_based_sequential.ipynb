```python
from langgraph.graph import StateGraph, START, END
from langchain_groq import ChatGroq
from dotenv import load_dotenv
from typing import TypedDict


load_dotenv()
```

```python
class usecase(TypedDict):
    question: str
    answer: str
```

```python
model = ChatGroq(model='llama-3.1-8b-instant')
```

```python
# def recieve_question(state: usecase) -> usecase:
#     question = input("Please enter your question: ")
#     state['question'] = question
#     return state
```

```python
def give_answer(state:usecase)-> usecase:
    question = state['question']
    answer = model.invoke(question).content
    state['answer'] = answer
    return state
```

```python
graph = StateGraph(usecase)
# create a graph with the state which is usecase here 
graph.add_node("Give Answer", give_answer)

# now add edges 
graph.add_edge(START,"Give Answer")
graph.add_edge("Give Answer",END)

workflow = graph.compile()
```

```python
initial_state = {'question':'Why an urn is the Ashes trophy?'}
final_answer = workflow.invoke(initial_state)
print(final_answer['answer'])
```
