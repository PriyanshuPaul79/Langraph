```python
from langgraph.graph import StateGraph, START, END
from langchain_groq import ChatGroq 
from dotenv import load_dotenv
from typing import TypedDict

load_dotenv()
```

```python
class PCStructure(TypedDict):
    instruction: str
    response: str
```

```python
model = ChatGroq(model='llama-3.1-8b-instant')
```

```python
def make_outline(state:PCStructure) -> PCStructure:
    task = state['instruction']
    prompt = f'create a basic outline in order to write a blog about {task}'
    answer = model.invoke(prompt).content
    state['response'] = answer 
    return state
```

```python
def make_blog(state:PCStructure)-> PCStructure:
    outline = state['response']
    prompt = f'you are given some outlines regarding {outline} write a short blog using these.'
    result = model.invoke(prompt).content
    state['response'] = result
    return state
```

```python
graph = StateGraph(PCStructure)

graph.add_node("make_outline",make_outline)
graph.add_node("make_blog",make_blog)

graph.add_edge(START,"make_outline")
graph.add_edge("make_outline","make_blog")
graph.add_edge("make_blog",END)

workflow = graph.compile()
```

```python
initial_state = {'instruction':'tell me something about Hanuman Chalisa'}
final_result = workflow.invoke(initial_state)
print(final_result['response'])
```
